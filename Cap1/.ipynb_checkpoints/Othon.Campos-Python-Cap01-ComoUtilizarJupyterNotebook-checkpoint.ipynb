## Graficos

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)
ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100)
plt.show()
```

```python
import numpy as np
import pandas as pd
#Gera gráfico na célula de baixo
%matplotlib inline
from matplotlib import pyplot as plt

#Ajusta o tamanho do gráfico
plt.rcParams['figure.figsize'] = (10,5)
#Eixo x
estados = ['SP', 'SE', 'BA', 'RJ', 'PR']
#Eixo y
num_vendas = [5,11,3,8,10]
xs = [i + 0.5 for i, _ in enumerate(estados)]
plt.bar(xs,num_vendas)
#Nome do eixo y
plt.ylabel("Qtde de Vendas")
#Título do gráfico
plt.title("Vendas de Carros por Estado")
plt.xticks([i + 0.5 for i, _ in enumerate(estados)], estados)

plt.show()
```
